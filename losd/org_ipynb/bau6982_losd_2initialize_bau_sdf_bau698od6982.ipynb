### Gruppierungen

```python
agg_jahr = data2betested.loc[data_min_date:data_max_date]\
    .groupby(['StichtagDatJahr','StichtagDatJahr_str', 'Jahr',])\
    .agg(
        sum_BEW_GGH8109=('BEW_GGH8109', 'sum'),sum_WHG_GGH8109=('WHG_GGH8109', 'sum')
        ,sum_BEW_WHF_GGH8109=('WHF_GGH8109', 'sum'), med_WFP=('WFP_GGH8109_STK0001', 'median')
        ,med_PPW=('PPW_GGH8109_STK0001', 'median')
    )\
    .sort_values('Jahr', ascending=False) 
agg_jahr.reset_index().head(3)
```

```python
agg_Quarts = data2betested.loc[data_min_date:data_max_date]\
    .groupby(['RAUM_LANG', 'RAUM_CODE']) \
    .agg(
        sum_BEW_GGH8109=('BEW_GGH8109', 'sum'),sum_WHG_GGH8109=('WHG_GGH8109', 'sum')
        ,sum_BEW_WHF_GGH8109=('WHF_GGH8109', 'sum'), med_WFP=('WFP_GGH8109_STK0001', 'median')
        ,med_PPW=('PPW_GGH8109_STK0001', 'median')
    )\
    .sort_values('RAUM_CODE', ascending=False) 
agg_Quarts.reset_index().head(34)
```

```python
agg_Kreis = data2betested.loc[data_min_date:data_max_date]\
    .groupby(['KREISEZH_LANG', 'KREISEZH_CODE',]) \
    .agg(
        sum_BEW_GGH8109=('BEW_GGH8109', 'sum'),sum_WHG_GGH8109=('WHG_GGH8109', 'sum')
        ,sum_BEW_WHF_GGH8109=('WHF_GGH8109', 'sum'), med_WFP=('WFP_GGH8109_STK0001', 'median')
        ,med_PPW=('PPW_GGH8109_STK0001', 'median')
    )\
    .sort_values('KREISEZH_CODE', ascending=True) 
agg_Kreis.reset_index().head(13)
```

```python
agg_Eig = data2betested.loc[data_min_date:data_max_date]\
    .groupby(['EIG_LANG', 'EIG_CODE', 'WGB_LANG','WGB_CODE',]) \
    .agg(
        sum_BEW_GGH8109=('BEW_GGH8109', 'sum'),sum_WHG_GGH8109=('WHG_GGH8109', 'sum')
        ,sum_BEW_WHF_GGH8109=('WHF_GGH8109', 'sum'), med_WFP=('WFP_GGH8109_STK0001', 'median')
        ,med_PPW=('PPW_GGH8109_STK0001', 'median')
    )\
    .sort_values('EIG_CODE', ascending=True) 
agg_Eig .reset_index().head(13)
```
